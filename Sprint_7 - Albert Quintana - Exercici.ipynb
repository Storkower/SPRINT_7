```python
def obtenir_entrada():
    """
    Sol·licita i valida les dades d'entrada per al pes, l'alçada i la cintura.
    Retorna el pes en kg, l'alçada en metres i la cintura en cm.
    """
    while True:
        try:
            pes = float(input("Introdueix el teu pes en kg: "))
            if pes <= 0:
                raise ValueError("El pes ha de ser un valor positiu.")
            break
        except ValueError as e:
            print(f"Entrada no vàlida: {e}. Torna-ho a intentar.")

    while True:
        try:
            alçada = float(input("Introdueix la teva alçada en metres: "))
            if alçada <= 0:
                raise ValueError("L'alçada ha de ser un valor positiu.")
            break
        except ValueError as e:
            print(f"Entrada no vàlida: {e}. Torna-ho a intentar.")
    
    while True:
        try:
            cintura = float(input("Introdueix la teva cintura en cm: "))
            if cintura <= 0:
                raise ValueError("La cintura ha de ser un valor positiu.")
            break
        except ValueError as e:
            print(f"Entrada no vàlida: {e}. Torna-ho a intentar.")

    return pes, alçada, cintura


def calcular_imc(pes, alçada):
    """
    Calcula l'IMC a partir del pes i l'alçada.
    """
    return pes / (alçada ** 2)


def calcular_cca(cintura, alçada):
    """
    Calcula la relació cintura/alçada (CCA).
    """
    alçada_cm = alçada * 100  # Convertim l'alçada a centímetres
    return cintura / alçada_cm


def classificar_imc(imc):
    """
    Classifica l'IMC segons les categories establertes al link de l'exercici 'Muysalud'.
    Alana em diu que és més òptim no fer servir trams i fer servir el nivell inferior. ( En comptes de 18.5 < imc <25 fer servir imc < 18.5 i imc < 25)
    """
    if imc < 18.5:
        return "Baix pes"
    elif imc < 25:
        return "Pes normal"
    elif imc < 30:
        return "Sobrepès"
    else:80


def mostrar_resultat(imc, categoria_imc, cca):
    """
    Mostra el resultat de l'IMC, la seva classificació, i la relació cintura/alçada (CCA).
    """
    print(f"El teu IMC és: {imc:.2f}")
    print(f"La teva categoria d'IMC és: {categoria_imc}")
    print(f"La teva relació cintura/alçada (CCA) és: {cca:.2f}")
    if cca > 0.5:
        print("Atenció: La teva relació cintura/alçada és superior a 0.5, indicant un major risc de malalties relacionades amb l'obesitat.")
    else:
        print("La teva relació cintura/alçada és dins dels límits saludables.")


def main():
    """
    Funció principal que realitza el procés de càlcul i classificació de l'IMC i la CCA.
    """
    pes, alçada, cintura = obtenir_entrada()
    imc = calcular_imc(pes, alçada)
    categoria_imc = classificar_imc(imc)
    cca = calcular_cca(cintura, alçada)
    mostrar_resultat(imc, categoria_imc, cca)


""" En aquest cas no és necessari el 'if __name__ == "__main__":'. Això només quan s'executen automàticament els fitxers python"""
main()
```

Convertidor de temperatures.

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions.

```python
# Constants per a les unitats de temperatura
UNITS = ["Celsius", "Fahrenheit", "Kelvin", "Rankine", "Réaumur"]

def convertir_temperatura(valor, des_de, a_on):
    """
    Converteix una temperatura entre dues unitats específiques.
    
    Arguments:
    - valor (float): la temperatura que volem convertir.
    - des_de (str): unitat d'origen.
    - a_on (str): unitat de destí.
    
    Retorna:
    - float: la temperatura convertida a la unitat de destí.
    
    Nota: La taula amb les fórmules ho he tret d'una imatge d'internet. No he verificat que la fórmula sigui exacte.
    """
    conversions = {
        ("Celsius", "Fahrenheit"): lambda x: (x * 9/5) + 32, 
        ("Celsius", "Kelvin"): lambda x: x + 273.15,
        ("Celsius", "Rankine"): lambda x: (x + 273.15) * 9/5, 
        ("Celsius", "Réaumur"): lambda x: x * 4/5, 
        
        ("Fahrenheit", "Celsius"): lambda x: (x - 32) * 5/9,
        ("Fahrenheit", "Kelvin"): lambda x: (x - 32) * 5/9 + 273.15,
        ("Fahrenheit", "Rankine"): lambda x: x + 459.67,
        ("Fahrenheit", "Réaumur"): lambda x: (x - 32) * 4/9,
        
        ("Kelvin", "Celsius"): lambda x: x - 273.15,
        ("Kelvin", "Fahrenheit"): lambda x: (x - 273.15) * 9/5 + 32,
        ("Kelvin", "Rankine"): lambda x: x * 9/5,
        ("Kelvin", "Réaumur"): lambda x: (x - 273.15) * 4/5,
        
        ("Rankine", "Celsius"): lambda x: (x - 491.67) * 5/9,
        ("Rankine", "Fahrenheit"): lambda x: x - 459.67,
        ("Rankine", "Kelvin"): lambda x: x * 5/9,
        ("Rankine", "Réaumur"): lambda x: (x - 491.67) * 4/9,
        
        ("Réaumur", "Celsius"): lambda x: x * 5/4,
        ("Réaumur", "Fahrenheit"): lambda x: (x * 9/4) + 32,
        ("Réaumur", "Kelvin"): lambda x: (x * 5/4) + 273.15,
        ("Réaumur", "Rankine"): lambda x: (x * 9/4) + 491.67,
    }

    # Validar la conversió
    try:
        return conversions[(des_de, a_on)](valor)
    except KeyError:
        raise ValueError(f"La conversió de {des_de} a {a_on} no és vàlida.")


def obtenir_unitat():
    """
    Sol·licita a l'usuari la unitat d'origen.
    
    Retorna:
    - str: la unitat d'origen seleccionada.
    """
    while True:
        unitat_origen = input(f"Introdueix la unitat d'origen ({', '.join(UNITS)}): ").capitalize()
        if unitat_origen in UNITS:
            return unitat_origen        
        print(f"Unitat d'origen no vàlida. Assegura't d'introduir una unitat correcta ({', '.join(UNITS)}).")


def obtenir_temperatura(unitat_origen):
    """
    Sol·licita a l'usuari la temperatura a convertir en una unitat d'origen específica.
    
    Arguments:
    - unitat_origen (str): unitat en què l'usuari introduirà la temperatura.
    
    Retorna:
    - float: la temperatura introduïda.
    """
    while True:
        try:
            temperatura = float(input(f"Introdueix la temperatura en {unitat_origen}: "))
            return temperatura
        except ValueError:
            print("Error: La temperatura no és vàlida. Torna-ho a intentar.")


def mostrar_resultat(temperatura, unitat_origen):
    """
    Mostra els resultats de la conversió a totes les unitats, excepte la unitat d'origen.
    
    Arguments:
    - temperatura (float): temperatura a convertir.
    - unitat_origen (str): unitat d'origen seleccionada.
    """
    unitats = UNITS.copy()
    unitats.remove(unitat_origen)  # Eliminem la unitat d'origen de la llista de destinació, no converteixes a la mateixa unitat

    print(f"\nConversió de {temperatura} {unitat_origen}:")
    for unitat_dest in unitats:
        try:
            resultat = convertir_temperatura(temperatura, unitat_origen, unitat_dest)
            print(f"{temperatura} {unitat_origen} és igual a {resultat:.2f} {unitat_dest}")
        except ValueError as e:
            print(e)


def main():
    """
    Funció principal que gestiona el flux de l'aplicació.
    """
    print("Benvingut al conversor de temperatures!")
    unitat_origen = obtenir_unitat()  # Primer es demana la unitat d'origen
    temperatura = obtenir_temperatura(unitat_origen)  # Després, es demana la temperatura
    mostrar_resultat(temperatura, unitat_origen)  # Finalment, es mostren les conversions


# Veure explicació exercici 1 nivell 1 
#if __name__ == "__main__":
main()
```

Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
"""
La classe Counter de la llibreria collections és una manera eficient i elegant de comptar elements en una llista o iterador. Ens permet comptar les paraules d'una manera més compacta i ràpida que utilitzant un diccionari manualment.
	Counter també torna un diccionari on les claus són les paraules i els valors són els comptes d'aquestes.
"""
import string
from collections import Counter

# Llista de paraules comuns a ignorar (stop words) | això ho feien servir a un vídeo de youtube. Volen totes a l'exercici
# STOP_WORDS = {"i", "el", "la", "de", "en", "a", "per", "que", "amb", "un", "una", "els", "les"}

def netejar_text(text):
    """
    Neteja el text eliminant signes de puntuació i convertint-lo a minúscules.

    Arguments:
    - text (str): El text a netejar.

    Retorna:
    - str: El text netejat, tot en minúscules i sense signes de puntuació. Podem també el·liminar accents i dièresi
    """
    # Convertir a minúscules
    text = text.lower()
    
    # Eliminar signes de puntuació
    text = text.translate(str.maketrans("", "", string.punctuation))
    
    # Eliminar accents i dièresi | tinc paraula amb i sense accent per provar-ho.
    # text = text.translate(str.maketrans("áàäéèëíìïòóöùúüÀÁÄÈÉËÌÍÏÒÓÖÙÚÜ", "aaaeeeiiiooouuuAAAEEEIIIOOOUUU"))

    return text


def comptar_paraules(text):
    """
    Comptar les vegades que apareix cada paraula en un text, excloent les paraules comunes si fem servir STOP_WORDS.

    Arguments:
    - text (str): El text a analitzar.

    Retorna:
    - dict: Un diccionari amb les paraules com a claus i el nombre de vegades que apareixen com a valors.
    """
    # Comprovació de tipus d'entrada | comprovem que sigui una cadena de caràcters (str) si no error
    if not isinstance(text, str):
        raise ValueError("El text ha de ser una cadena de caràcters.")

    # Netejar el text
    text = netejar_text(text)
    
    # Dividir el text en paraules
    paraules = text.split()
    
    # Filtrar les paraules comunes (stop words)
    # paraules_filtrades = [paraula for paraula in paraules] if paraula not in STOP_WORDS] | No ho estem fent servir
    
    # Comptar les paraules utilitzant Counter
    comptador = Counter(paraules)
    
    return dict(comptador)


def mostrar_resultat(comptador):
    """
    Mostra les paraules i el nombre de vegades que apareixen.

    Arguments:
    - comptador (dict): Diccionari amb les paraules i el nombre de vegades que apareixen.
    """
    print("\nResultat de la comptabilització de paraules:")
    for paraula, compte in comptador.items():
        print(f"'{paraula}': {compte}")

# Exemple de text per veure com funciona. Hauria de fer-se dinàmic introduïnt el text cada cop que es vol fer servir
# Veure explicació exercici 1 nivell 1 'if __name__ == "__main__":'
# if __name__ == "__main__":
#  text = """Python és un llenguatge de programació. Python és molt popular, i és utilitzat per moltes persones aquest llenguatge de programacio."""

# S'ha afegit una gestió bàsica d'errors utilitzant try-except per tractar casos en què l'entrada no sigui un text vàlid (per exemple, si es passa un nombre en lloc d'una cadena).
def main_comptar():
    text = input('Ingresa el texto')
    try:
        resultat = comptar_paraules(text)
        mostrar_resultat(resultat)
    except ValueError as e:
        print(f"Error: {e}")
```

Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.
    reverse_dictionary({'a': 1, 'b': 2, 'c': 3,})
{1: 'a', 2: 'b', 3: 'c'}
    reverse_dictionary({'x': 'apple', 'y': 'banana', 'z': 'banana'})
Error: multiple keys for one value

```python
"""El diccionari reversed_dict és un diccionari estàndard de Python, que ofereix un accés ràpid per
 clau (O(1) en mitjana) i és eficient per manejar valors duplicats sense necessitat de múltiples 
 recorreguts."""

def reverse_dictionary(d):
    # Comprovem si els valors són únics abans de començar a invertir
    """La línia if len(d.values()) != len(set(d.values())): compara la longitud de la llista 
    de valors del diccionari d amb la longitud del conjunt (set(d.values())). Un conjunt elimina 
    els duplicats, així que si la longitud no coincideix, vol dir que hi ha valors duplicats."""
    
    if len(d.values()) != len(set(d.values())):
        # Si hi ha valors duplicats, imprimeix l'error i surt de la funció
        print("Error: multiple keys for one value")
        return  # Sortim de la funció sense retornar cap valor

    # Si els valors són únics, invertim el diccionari utilitzant diccionari de comprensió
    reversed_dict = {value: key for key, value in d.items()}
    
    return reversed_dict

# Veure com funciona
resultat_1 = reverse_dictionary({'a': 1, 'b': 2, 'c': 3})  
print(resultat_1)  # Sortida esperada: {1: 'a', 2: 'b', 3: 'c'}

resultat_2 = reverse_dictionary({'x': 'apple', 'y': 'banana', 'z': 'banana'})
# print(resultat_2)  # Sortida esperada: Error: multiple keys for one value (i no retorna None)
# No caldrà imprimir 'None', ja que el retorn de la funció no es mostra
```

Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.

    reverse_dictionary_plus({'x': 'apple', 'y': 'banana', 'z': 'banana'})
{'apple': 'x', 'banana': ['y', 'z']}

```python
def reverse_dictionary_plus(d):
    # Diccionari buit per a emmagatzemar el resultat
    reversed_dict = {} # veure explicació exercici anterior

    # Recorrer cada clau, valor del diccionari original
    for key, value in d.items():
        # Si el valor ja existeix com a clau al diccionari invertit
        if value in reversed_dict:
            # Si és únic, convertim a llista
            if isinstance(reversed_dict[value], list):
                reversed_dict[value].append(key)
            else:
                reversed_dict[value] = [reversed_dict[value], key]
        else:
            # Si no existeix, l'assignem com a valor únic
            reversed_dict[value] = key

    return reversed_dict

# Veure com funciona
resultat_1 = reverse_dictionary_plus({'x': 'apple', 'y': 'banana', 'z': 'banana'})
print(resultat_1)  # Sortida esperada: {'apple': 'x', 'banana': ['y', 'z']}

# les lletres i els números eren del anterior
# resultat_2 = reverse_dictionary_plus({'a': 1, 'b': 2, 'c': 3})
# print(resultat_2)  # Sortida esperada: {1: 'a', 2: 'b', 3: 'c'}
```

Conversió de tipus de dades

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir. Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

conversion(['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1,2, '3', '3.4']])

([1.3, 10000000000.0, 2.0, 1.0, 1.4, 1.0, 2.0, 3.0, 3.4],
['one', 'seven', '3-1/2', 'not-a-number'])

```python
def conversion(data):
    float_list = []  # Llista per emmagatzemar els elements que es poden convertir a flotants
    non_float_list = []  # Llista per emmagatzemar els elements que no es poden convertir a flotants
    
    for item in data:
        # Si l'element és un string, intentem convertir-lo
        if isinstance(item, str):
            # mètode de control d'errors
            try:
                # Intentem convertir el string a un número flotant amb float(item), si té éxit ho afegim a float_list
                float_list.append(float(item))
            except ValueError:
                # Si no és possible, l'afegim a la llista d'elements no convertibles
                non_float_list.append(item)
        # Si l'element és un tuple o una llista, recorrem els seus elements i intentem convertir-los
        elif isinstance(item, (tuple, list)):
            for sub_item in item:
                try:
                    # Intentem convertir els sub-elements a flotants
                    float_list.append(float(sub_item))
                except ValueError:
                    # Si no és possible, l'afegim a la llista d'elements no convertibles
                    non_float_list.append(sub_item)
        else:
            # Si l'element no és string, tuple ni llista, no podem convertir-lo a flotant directament
            non_float_list.append(item)
    
    return float_list, non_float_list
# Veure com funciona
resultat = conversion(['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']])
print(resultat)
# Sortida esperada: 
# ([1.3, 10000000000.0, 2.0, 1.0, 1.4, 1.0, 2.0, 3.0, 3.4], ['one', 'seven', '3-1/2', 'not-a-number'])
```

Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
# string: Utilitzem aquesta llibreria per obtenir l'alfabet de lletres (A-Z) per ordenar les claus del diccionari.
import string
# defaultdict: Una estructura de dades de Python que permet crear diccionaris amb valors per defecte (en aquest cas, un altre diccionari per les paraules). Ens ajuda a no haver de crear manualment el diccionari de subclausules (per a cada lletra).
from collections import defaultdict
# re: La llibreria per expressions regulars, que utilitzem per separar les paraules del text.
import re

def comptar_i_endreçar_paraules(archivo):
    # Diccionari per comptar les paraules per lletra inicial
    paraules_per_lletra = defaultdict(lambda: defaultdict(int))
    
    # Carregar el conjunt de paraules vàlides
    # Podries crear un conjunt de paraules vàlides per filtrar paraules incorrectes, però ara no l'utilitzem
    # Exemples de paraules que podríem filtrar: 'vate', 'vea', etc.
    
    # Obrim l'arxiu | mode lectura 'r' i 'utf-8' per garantir llegeix accents, ç
    with open(archivo, 'r', encoding='utf-8') as file:
        # Llegim el contingut del fitxer
        text = file.read().lower()  # Convertim tot a minúscules
        
        # Usar expressió regular per separar paraules només
        # La regex captura paraules compostes per lletres i apòstrofs o guions (ex: l'aigua)
        paraules = re.findall(r'\b[a-zà-ÿÀ-ÿ\'-]+\b', text)
        
        # Afegim cada paraula a la llista de la seva lletra inicial i comptem les ocurrències
        for paraula in paraules:
            if paraula:  # Per evitar paraules buides
                # Validació per evitar paraules no vàlides com 'vate', 'vea', etc.
                if len(paraula) >= 1 and paraula.isalpha():  # Si no és una paraula alfabètica pura, la ignorem
                    primera_lletra = paraula[0]
                    paraules_per_lletra[primera_lletra][paraula] += 1
    
    # Creem el diccionari final ordenat
    resultat_ordenat = {}
    
    # Ordenem les claus per lletres de la A a la Z
    for lletra in string.ascii_lowercase: #  string.ascii_lowercase per obtenir les lletres de l'alfabet de la A a la Z.
        if lletra in paraules_per_lletra:
            # Ordenem les paraules alfabèticament dins de cada lletra
            paraules_ordenades = dict(sorted(paraules_per_lletra[lletra].items())) # Convertim les parelles de paraula i freqüència a un diccionari ordenat.
            # Afegim al diccionari resultat
            resultat_ordenat[lletra] = paraules_ordenades
    
    return resultat_ordenat

def main_dict(ruta_arxiu):
    # Cridem la funció amb la ruta del fitxer
    try:
        resultat = comptar_i_endreçar_paraules(ruta_arxiu)

        # Imprimim el resultat en el format desitjat (diccionari ordenat per lletres i paraules)
        print(resultat)
    except FileNotFoundError:
        print(f"Error: El fitxer {ruta_arxiu} no es troba.")
    except Exception as e:
        print(f"S'ha produït un error: {e}")
```
